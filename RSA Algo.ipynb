```python
import random
import sympy
```

```python
def gcd(a,b):
  while b!=0:
    a,b = b,a%b
  return a

def find_e(phi):
  e = 2
  while e<phi:
    if gcd(e,phi)==1:
      return e
    else:
      e = e+1

def find_d(e, phi):
  d = 2
  while d<phi:
    if((d*e)%phi) == 1:
      return d
    else:
      d = d+1
```

```python
def generate_keypair(p,q):
  n = p*q
  phi = ((p-1)*(q-1))

  e = find_e(phi)

  d = find_d(e,phi)

  return((e,n),(d,n))
```

```python
def encrypt(pk, plaintext):
    key, n = pk
    cipher = [(ord(char)**key)%n for char in plaintext]
    return cipher
```

```python
def decrypt(pk, ciphertext):
    key, n = pk
    plain = [chr((char**key)%n) for char in ciphertext]
    return ''.join(plain)
```

```python
p = sympy.randprime(1,100)
q = sympy.randprime(1,100)

public,private = generate_keypair(p,q)

message = input("Enter message: ")

encrypted_msg = encrypt(public, message)

print(encrypted_msg)

decrypted_msg = decrypt(private, encrypted_msg)

print(decrypted_msg)
```
